```python
import pandas as pd
import numpy as np
import warnings; warnings.simplefilter('ignore')
SYMBOL_STOCK = "AXISBANK"
SYMBOL_NEWS = "AXISBANK"
```

```python
text = pd.read_csv("../data/"+SYMBOL_STOCK+'_news.csv')
# text.drop(text.columns[[0, 5, 6]], axis = 1, inplace=True)
text['time'] = (text['year']-2000)*365 + text['month']*30 + text['date']
# text.head()
```

```python
stock = pd.read_csv("../data/"+SYMBOL_STOCK+'_stocks.csv')

# Here 1 indicates that the stock would go up as close - open is a positive value

stock['year']=pd.to_numeric(stock['Date'].str.slice(0,4))
stock['month']=pd.to_numeric(stock['Date'].str.slice(5,7))
stock['date']=pd.to_numeric(stock['Date'].str.slice(8,10))
stock['time'] = (stock['year']-2000)*365 + stock['month']*30 + stock['date']

stock.loc[stock.Close > stock.Open, 'label'] = "1"
stock.loc[stock.Close <= stock.Open, 'label'] = "0"
stock.set_index(stock.time, inplace=True)
stock.sample(5)
```

```python
X = text.loc[text.time == text.time.max(), ['title']]
X.to_csv("../data/"+SYMBOL_STOCK+'Test.csv',index = True)
print(X)
```

```python
for uni in text.time.unique():
    try:
        text.loc[text.time == uni, 'label'] = stock.loc[uni, 'label']
    except KeyError:
        text.loc[text.time == uni, 'label'] = np.nan
text.dropna(axis=0, how='any', inplace=True)
text.drop(text.columns[[0,1,3,4,5,6,7]], axis = 1, inplace=True)
text.sample(5)

text.to_csv("../data/"+SYMBOL_STOCK+'.csv',index = True)
```
